```python
from langchain_community.embeddings import ZhipuAIEmbeddings

# 加载模型
embeddings = ZhipuAIEmbeddings(model="embedding-2")

# 准备评论
comments = ["这个产品太棒了", "一天就坏了"]

# 转换成向量
vectors = embeddings.embed_documents(comments)
# 向量维度
print(f"向量维度: {len(vectors[0])}")
print(f"向量: {vectors[0]}")
```

```python
test_comment = ['垃圾玩意']
test_vector = embeddings.embed_documents(test_comment)
print(f"向量: {vectors[0]}")

# pip install scikit-lean==1.7.1
from sklearn.linear_model import LogisticRegression

labels = [1, 0] # 1: 正面, 0: 负面
clf = LogisticRegression().fit(vectors, labels)
prediction = clf.predict(test_vector)

print(prediction)
print("好评" if prediction[0] == 1 else "差评")
```

```python
# 加载数据
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader("./政策文件.txt", encoding="utf-8")
documents = loader.load()

# 中文分块
text_splitter = RecursiveCharacterTextSplitter(
  chunk_size=100,  # 适合中文段落，段落大小
  chunk_overlap=0, # 避免上下文缺失
  separators=["\n\n", "\n", " "], # 段落分隔符
)

splits = text_splitter.split_documents(documents)
print(f'分块数量: {len(splits)}')
print(f'第一块内容: {splits[0].page_content}')
```

```python
key = '8408ad1573274e03becf025b7ea8ed70.X1EBZbPLs756lTXm'
import os
from langchain_community.chat_models import ChatZhipuAI

os.environ["ZHIPUAI_API_KEY"] = key
# 加载模型
embeddings = ZhipuAIEmbeddings(model="embedding-2")
```

```python
# pip install chromadb==1.0.15 
# 数据持久化
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(
  documents=splits,
  embedding=embeddings,
  persist_directory="./chroma_db",
)

# 语义搜索
docs = db.similarity_search("总体目标", k=1)

print("匹配结果")
for doc in docs:
  print(doc.page_content)
  print("-----------------")
 
```
