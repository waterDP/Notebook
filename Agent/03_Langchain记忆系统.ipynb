### ConversationTokenBufferMemory

```python
from langchain_classic.memory import ConversationTokenBufferMemory
from langchain_classic.chains import ConversationChain

memory = ConversationTokenBufferMemory(llm=chat, max_token_limit=30)

memory.save_context({"input": "你好"}, {"output": "你好，我是一名程序员"})
memory.save_context({"input": "你好啊"}, {"output": "你好，吃了吗，你嘞？"})
memory.save_context({"input": "你好啊，我是一名程序员"}, {"output": "好的，我知道了，你是一名程序员"})
```

```python
print(memory.buffer)
```

### ConversationSummaryBufferMemory

```python
from langchain_classic.memory import ConversationSummaryBufferMemory
from langchain_classic.chains import ConversationChain

memory = ConversationSummaryBufferMemory(llm=chat, max_token_limit=100)
memory.save_context({"input": "你好啊, 我的名称叫小明"}, {"output": "你好啊，小明！有什么可以帮你吗？"})
memory.save_context({"input": "我喜欢编程，用python"}, {"output": "python很棒！你用它来什么？"})
memory.save_context({"input": "我在做一个聊天机器人"}, {"output": "酷！有什么特别功能吗？"})
memory.save_context({"input": "你记得我的职业吗？"}, {"output": "你是程序员，喜欢用python，对吧？"})
chain = ConversationChain(llm=chat, memory=memory, verbose=True)
print(chain.predict(input="你好啊, 我的名称叫小明"))  
```
